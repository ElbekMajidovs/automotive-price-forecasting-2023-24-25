```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
file_path1 = Path('../data/raw/autoscout24_2021.csv')
file_path2 = Path('../data/raw/autoscout24_2022.csv')
file_path3 = Path('../data/raw/autoscout24_2023.csv')
file_path4 = Path('../data/raw/autoscout24_2024.csv')
file_path5 = Path('../data/raw/autoscout24_full.csv')
```

```python
import pandas as pd
import numpy as np

# --- helper: kW → HP ---
def kw_to_hp(x):
    return x * 1.341 if pd.notnull(x) else np.nan

# =======================
# DF1
# =======================
df1_ = df1.rename(columns={
    "Vehicle_brand": "make",
    "Vehicle_model": "model",
    "Price": "price",
    "Production_year": "year",
    "Mileage_km": "mileage",
    "Fuel_type": "fuel",
    "Transmission": "transmission",
    "Power_HP": "power_hp",
    "Displacement_cm3": "engine_cc",
    "Doors_number": "doors",
    "country": "country"
})

df1_ = df1_[["make","model","price","year","mileage","fuel",
             "transmission","power_hp","engine_cc","doors","country"]]


# =======================
# DF2
# =======================
df2_ = df2.rename(columns={
    "make": "make",
    "model": "model",
    "price": "price",
    "first_registration": "year",
    "mileage": "mileage",
    "fuel_type": "fuel",
    "gearbox": "transmission",
    "power": "power_hp",
    "engine_size": "engine_cc",
    "doors": "doors",
    "country": "country"
})

df2_ = df2_[["make","model","price","year","mileage","fuel",
             "transmission","power_hp","engine_cc","doors","country"]]


# =======================
# DF3
# =======================
df3_ = df3.rename(columns={
    "Brand": "make",
    "ModelC": "model",
    "Price": "price",
    "Année": "year",
    "Kilométrage": "mileage",
    "Carburant": "fuel",
    "Transmission": "transmission",
    "Puissance kW": "power_kw",
    "Portes": "doors",
    "Country": "country"
})

df3_["power_hp"] = df3_["power_kw"].apply(kw_to_hp)
df3_["engine_cc"] = np.nan

df3_ = df3_[["make","model","price","year","mileage","fuel",
             "transmission","power_hp","engine_cc","doors","country"]]


# =======================
# DF4
# =======================
df4_ = df4.rename(columns={
    "Make": "make",
    "Model": "model",
    "Price": "price",
    "Year": "year",
    "Mileage_km": "mileage",
    "Fuel_Type": "fuel",
    "Gearbox": "transmission",
    "Power_hp": "power_hp",
    "Engine_Size_cc": "engine_cc",
    "Doors": "doors",
    "Country": "country"
})

df4_ = df4_[["make","model","price","year","mileage","fuel",
             "transmission","power_hp","engine_cc","doors","country"]]


# =======================
# DF5
# =======================
df5_ = df5.rename(columns={
    "make": "make",
    "model": "model",
    "price": "price",
    "production_year": "year",
    "mileage_km_raw": "mileage",
    "fuel_category": "fuel",
    "transmission": "transmission",
    "power_hp": "power_hp",
    "cylinders_volume_cc": "engine_cc",
    "nr_doors": "doors",
    "country_code": "country"
})

df5_ = df5_[["make","model","price","year","mileage","fuel",
             "transmission","power_hp","engine_cc","doors","country"]]


# =======================
# CONCAT ALL
# =======================
final_df = pd.concat([df1_, df2_, df3_, df4_, df5_], ignore_index=True)


# =======================
# CLEANING (optional but recommended)
# =======================
# normalize text
final_df["make"] = final_df["make"].str.lower()
final_df["model"] = final_df["model"].str.lower()
final_df["fuel"] = final_df["fuel"].str.lower()
final_df["transmission"] = final_df["transmission"].str.lower()

# convert types
final_df["year"] = pd.to_numeric(final_df["year"], errors="coerce")
final_df["price"] = pd.to_numeric(final_df["price"], errors="coerce")
final_df["mileage"] = pd.to_numeric(final_df["mileage"], errors="coerce")

# drop obvious garbage rows
final_df = final_df.dropna(subset=["price", "make", "model"])

# reset index
final_df = final_df.reset_index(drop=True)

print(final_df.shape)
final_df.head()
```

```python
from sklearn.model_selection import train_test_split

# 60% train, 40% temp
df_train, df_temp = train_test_split(df, test_size=0.4, random_state=123)

# split temp into 20% valid, 20% test
df_valid, df_test = train_test_split(df_temp, test_size=0.5, random_state=123)
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error

def evaluate_model(model, X, y):
    y_pred = model.predict(X)
    
    mae = mean_absolute_error(y, y_pred)
    rmse = np.sqrt(mean_squared_error(y, y_pred))
    
    # avoid division by zero
    mape = np.mean(np.abs((y - y_pred) / np.clip(y, 1e-8, None))) * 100
    smape = np.mean(2 * np.abs(y - y_pred) / (np.abs(y) + np.abs(y_pred) + 1e-8)) * 100
    
    print(f"MAE  : {mae:.2f}")
    print(f"RMSE : {rmse:.2f}")
    print(f"MAPE : {mape:.2f}%")
    print(f"SMAPE: {smape:.2f}%")
    
    return {"mae": mae, "rmse": rmse, "mape": mape, "smape": smape}
```

## Train Dataset

```python
X_train = df_train.drop(columns=['price'])
y_train = df_train['price']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer

num_cols = X_train.select_dtypes(include=['float64','int64']).columns
cat_cols = X_train.select_dtypes(include=['object']).columns

preprocessor = ColumnTransformer([
    ('num', SimpleImputer(strategy='median'), num_cols),
    ('cat', Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', OneHotEncoder(handle_unknown='ignore'))
    ]), cat_cols)
])
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV

xgb_pipeline = Pipeline([
    ('prep', preprocessor),
    ('model', XGBRegressor(
        objective='reg:squarederror',
        tree_method='hist',
        n_jobs=-1,
        random_state=123
    ))
])

param_grid = {
    'model__n_estimators': [300, 500],
    'model__learning_rate': [0.03, 0.1],
    'model__max_depth': [4, 6],
    'model__subsample': [0.8, 1.0],
    'model__colsample_bytree': [0.8, 1.0],
    'model__min_child_weight': [1, 5],
    'model__gamma': [0, 0.5],
    'model__reg_alpha': [0, 0.01],
    'model__reg_lambda': [1, 3]
}

xgb_grid_search = GridSearchCV(
    xgb_pipeline,
    param_grid,
    scoring='neg_mean_absolute_error',
    cv=3,
    n_jobs=-1,
    verbose=1
)

xgb_grid_search.fit(X_train, y_train)

model_xgb_best = xgb_grid_search.best_estimator_

print("Best params:", xgb_grid_search.best_params_)
print("Best MAE:", -xgb_grid_search.best_score_)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline

gbr_pipeline = Pipeline([
    ('prep', preprocessor),
    ('model', GradientBoostingRegressor(random_state=123))
])

param_grid = {
    'model__n_estimators': [200, 300],
    'model__learning_rate': [0.03, 0.1],
    'model__max_depth': [3, 4],
    'model__min_samples_split': [2, 10],
    'model__min_samples_leaf': [1, 2],
    'model__subsample': [0.8, 1.0]
}

gbr_grid_search = GridSearchCV(
    gbr_pipeline,
    param_grid,
    scoring='neg_mean_absolute_error',
    cv=3,
    n_jobs=-1,
    verbose=1
)

gbr_grid_search.fit(X_train, y_train)

model_gbr_best = gbr_grid_search.best_estimator_

print("Best params:", gbr_grid_search.best_params_)
print("Best MAE:", -gbr_grid_search.best_score_)
```

```python
from sklearn.ensemble import StackingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV

stack_pipeline = Pipeline([
    ('prep', preprocessor),
    ('model', StackingRegressor(
        estimators=[
            ('gbr', model_gbr_best.named_steps['model']),
            ('xgb', model_xgb_best.named_steps['model'])
        ],
        final_estimator=LinearRegression(),
        n_jobs=-1,
        passthrough=False
    ))
])

param_grid = {
    'model__final_estimator__fit_intercept': [True, False]
}

stack_grid_search = GridSearchCV(
    stack_pipeline,
    param_grid,
    scoring='neg_mean_absolute_error',
    cv=3,
    n_jobs=-1,
    verbose=1
)

stack_grid_search.fit(X_train, y_train)

model_stacking_best = stack_grid_search.best_estimator_

print("Best params:", stack_grid_search.best_params_)
print("Best MAE:", -stack_grid_search.best_score_)
```
